### **Adjusting the spacing around subplots**

```python
# An example that shrink the spacing of the subplots to zero
# change the value of wspace and hspace
# sharex=True parameter means that the subplots will share the same x-axis
# The "fig" variable is a reference to the entire figure
# The "axes" is a 2D array of subplot references
fig, axes = plt.subplots(2,2,sharex=True, sharey=False) # we have 2x2 sub plots

# The outer loop (for i in range(2)) iterates over rows
# The inner loop (for j in range(2)) iterates over columns
# Top left -> Top right -> bottom left -> bottom right
# (0, 0) = top left
# (0, 1) = top right
# (1, 0) = bottom left
# (1, 1) = bottom right
#Use for loop
for i in range(2):
    for j in range(2):
        axes[i,j].hist(rng.standard_normal(500), bins=50, color='navy', alpha=.5)

# Adjust the space around the subplot
# wspace = width of the subplots
# hspace = height of the padding
fig.subplots_adjust(wspace = .5, hspace = .1)
```

# **Adding shapes to a plot**
- <font color ="red">**Important for final exam**

```python
# Adding shapes to a plot
# always use the code for creating subplots
fig, ax=plt.subplots()

# Create a rectangle shape
# .4 -> width of the x axis
# .2 -> .75 + .2 = .95 -> height of the rectange
rect = plt.Rectangle((.2,.75), .4, .2, color='#800080', alpha=.8)


# Create a circle shape
# .15 -> radius of the circle
circ = plt.Circle([.7, .2], .15, color='#adff2f', alpha = .3)
# Create a triangle shape
# hexagon - > polygon of 6 sides
trig = plt.Polygon([[.15,.15], [.35, .4], [.2,.6]], color='#663399', alpha=.5)
# All all together
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(trig)
```

```python
# Adding shapes to a plot : Attention
import matplotlib.patches as mpatches
fig, ax = plt.subplots()

rect = plt.Rectangle((.2,.75), .4, .2, color='#556b2f', alpha=.8)
circ = plt.Circle([.7, .2], .15, color='#adff2f', alpha = .3)
tri = plt.Polygon([[.15,.15], [.35, .4], [.2,.6]], color='#663399', alpha=.5)

# Create a square shape
sqr = plt.Rectangle((.3,.1),.2,.2, color = '#ffff00', alpha = .9)

# Create an ellipse shape
# (0.5, 0.5) -> This tuple specifies the center of the ellipse
# 0.2 -> width of the ellipse
# 0.1 -> height of the ellipse
ellipse = mpatches.Ellipse((.5,.5), .2,.1, color='#00008b')
# add hexagon
# 6 -> specify number of sides in the regular polygon
# 0.1 -> distance from the center of the circumcircle (the point at the center of the hexagon) to any of its vertices
hex = mpatches.RegularPolygon((.7,.8), 6, .1, color='#db7093')

# Add all the subplot together
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(tri)
ax.add_patch(sqr)
ax.add_patch(ellipse)
ax.add_patch(hex);
```

# **Dealing with Data Frame: Important for final exam**

```python
# A little project: dataset about restaurant tipping
# Download the tips data here: https://bit.ly/3jCROII
tips = pd.read_csv("https://bit.ly/3jCROII")
tips
```

```python
# Using the crosstab() function
# To compute a simple cross-tabulation of two (or more) factors.
# Create a two factors data frame
party_counts = pd.crosstab(tips['day'], tips['size'])
party_counts
```

```python
# Reindex the DataFrame to the desidered Day order
desired_order = ['Thur', 'Fri', 'Sat', 'Sun']
party_counts_ordered = party_counts.reindex(desired_order)
party_counts_ordered
```

```python
# Extract party_counts from col 2 to 5
# Using loc method - explicit, using it as a label
# '2:5' is the name of the columns; they are not integer in this case
party_counts_ordered_edit = party_counts_ordered.loc[:,2:5]
party_counts_ordered_edit
```

```python
# Normalize each row to sum of 1
# axis="index" argument indicates that the division should be done along the rows
# normalizing the values in each row of the DataFrame party_counts_edit
# by dividing them by the sum of values in the respective row
party_pcts = party_counts_ordered_edit.div(party_counts_ordered_edit.sum(axis=1), axis ='index')
party_pcts
```

```python
# Example dataset
# Download the dataset here: https://bit.ly/3FYxCYZ
# Go to this link if you want to know more about this data: http://tinyurl.com/mr3zy799
macro = pd.read_csv("https://bit.ly/3FYxCYZ")
macro
```

```python
# Need to do normalisation; standardized the data
# Log transform the data
# diff() -> calculates the difference between that element and the previous one
trans_dat = np.log(dat).diff().dropna()
trans_dat
```

# **Before we go, a complex graph to end the day and the year :)**


```python
# Importing the datetime module
from datetime import datetime

# Create a figure and an axis
fig, ax = plt.subplots()

# Download the closing S&P 500 index price: 1990-02-01 until 2011-10-14
# The link: https://bit.ly/3iofIY8
data = pd.read_csv("https://bit.ly/3iofIY8", index_col=0, parse_dates=True)

# Extract the SPX data
spx = data["SPX"]

# Plot the SPX data from 1990 until 2011
spx.plot(ax=ax, color="blue")

# 2008–2009 financial crisis
crisis_data = [
    (datetime(2007, 10, 11), "Peak of bull market"),
    (datetime(2008, 3, 12), "Bear Stearns Fails"),
    (datetime(2008, 9, 15), "Lehman Bankruptcy")
]

# Using Matplotlib's annotate method to add annotations
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75), # Coordinate of the points
                xytext=(date, spx.asof(date) + 225), # Coordinate of the labels
                arrowprops=dict(facecolor="black", headwidth=4, width=2, headlength=4), # A dictionary specifying arrow properties
                horizontalalignment="left",
                verticalalignment="top")

# Zoom in on 2007-2010
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])

ax.set_title("Important dates in the 2008–2009 financial crisis");
```
